```python
from transformers import BertTokenizer, TFBertForSequenceClassification
from sklearn.model_selection import train_test_split
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('alldata_1_for_kaggle.csv', encoding='latin-1')
data.head()
```

```python
# Drop the first column
data = data.iloc[:, 1:]  # Keep all columns except the first

# Rename the new first and second columns
data.rename(columns={data.columns[0]: "class", data.columns[1]: "paper"}, inplace=True)

# Display the updated DataFrame
print(data.head())
```

```python
# Shape of the dataset
print("Shape of dataset:", data.shape)

# Summary statistics of numerical columns
print(data.describe())
```

```python
# Keep only unique rows based on the 'paper' column
data = data.drop_duplicates(subset='paper', keep='first')

# Display the resulting dataset
print(f"Shape of unique dataset: {data .shape}")
print(data .head())
```

```python
class_counts = data['class'].value_counts()

# Plot the histogram
plt.figure(figsize=(8, 6))
class_counts.plot(kind='bar', color='skyblue')
plt.title('Frequency of Classes')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

bars = plt.bar(class_counts.index, class_counts.values, color='skyblue')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             str(bar.get_height()), ha='center', va='bottom', fontsize=10)
plt.show()
```

```python
from sklearn.utils import resample
# Find the minimum class size
min_class_size = data['class'].value_counts().min()

# Balance the dataset by undersampling
balanced_data = data.groupby('class', group_keys=False).apply(
    lambda x: x.sample(30, random_state=42)
)

# Check the class distribution
print(balanced_data['class'].value_counts())
```

```python

# Extract the text and labels from the dataset
texts = balanced_data['paper']
labels = balanced_data['class'].astype('category').cat.codes # Convert classes to numeric codes

# Split the data: 80% for training, 20% for testing
train_texts, test_texts, train_labels, test_labels = train_test_split(
    texts, labels, test_size=0.2, random_state=42
)
```

```python
# Define the tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the text data
train_encodings = tokenizer(train_texts.tolist(), truncation=True, padding=True, max_length = 200)
test_encodings = tokenizer(test_texts.tolist(), truncation=True, padding=True, max_length = 200)
```

```python
num_classes = len(balanced_data['class'].unique())
train_labels_onehot = tf.keras.utils.to_categorical(train_labels, num_classes = num_classes)
test_labels_onehot = tf.keras.utils.to_categorical(test_labels, num_classes = num_classes)

# Convert the encodings to TensorFlow tensors
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    train_labels_onehot
))
test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    test_labels_onehot
))
```

```python
# Load BERT model with a classification head (adjust num_labels for the number of classes)
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_classes)
```

```python
# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=5e-5),
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Train the model
history = model.fit(
    train_dataset.batch(16), # Adjust batch size as needed
    validation_data=test_dataset.batch(16),
    epochs=3
)
```

```python
# Evaluate the model on the test dataset
test_loss, test_accuracy = model.evaluate(test_dataset.batch(16))

print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
from transformers import TFBertForSequenceClassification
import tensorflow as tf
import pandas as pd
import time

# Define hyperparameter grid
learning_rates = [3e-5, 5e-5]
batch_sizes = [8, 16]
epochs_list = [3, 5]
max_lengths = [100, 200]
# Store results
results = []

# Loop through each combination
for lr in learning_rates:
    for batch_size in batch_sizes:
        for epochs in epochs_list:
            for max_length in max_lengths:  # Include max_length in the loop
                print(f"Testing: LR={lr}, Batch Size={batch_size}, Epochs={epochs}, Max Length={max_length}")

                # Start timer
                start_time = time.time()

                # Tokenize data with current max_length
                train_encodings = tokenizer(train_texts.tolist(), truncation=True, padding=True, max_length=max_length)
                test_encodings = tokenizer(test_texts.tolist(), truncation=True, padding=True, max_length=max_length)

                # Prepare datasets
                train_dataset = tf.data.Dataset.from_tensor_slices((
                    {'input_ids': train_encodings['input_ids'], 'attention_mask': train_encodings['attention_mask']},
                    tf.keras.utils.to_categorical(train_labels, num_classes=num_classes)
                )).batch(batch_size)

                test_dataset = tf.data.Dataset.from_tensor_slices((
                    {'input_ids': test_encodings['input_ids'], 'attention_mask': test_encodings['attention_mask']},
                    tf.keras.utils.to_categorical(test_labels, num_classes=num_classes)
                )).batch(batch_size)

                # Load BERT model
                model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_classes)

                # Compile the model
                model.compile(
                    optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
                    loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
                    metrics=['accuracy']
                )

                # Train the model
                history = model.fit(
                    train_dataset,
                    validation_data=test_dataset,
                    epochs=epochs,
                    verbose=1
                )

                # Evaluate the model
                test_loss, test_accuracy = model.evaluate(test_dataset)

                # End timer
                end_time = time.time()
                elapsed_time = end_time - start_time

                # Display results for the combination
                print(f"Test Loss: {test_loss:.4f}")
                print(f"Test Accuracy: {test_accuracy:.4f}")
                print(f"Elapsed Time: {elapsed_time:.2f} seconds")

                # Save results
                results.append({
                    'learning_rate': lr,
                    'batch_size': batch_size,
                    'epochs': epochs,
                    'max_length': max_length,
                    'test_loss': test_loss,
                    'test_accuracy': test_accuracy,
                    'elapsed_time_seconds': elapsed_time
                })
```

```python
# Convert results to a DataFrame
results_df = pd.DataFrame(results)
results_df['elapsed_time_minutes'] = results_df['elapsed_time_seconds'] / 60

# Sort by accuracy
best_results = results_df.sort_values(by='test_accuracy', ascending=False)

# Display the top results
print("Top Hyperparameter Combinations:")
print(best_results.head())
```
